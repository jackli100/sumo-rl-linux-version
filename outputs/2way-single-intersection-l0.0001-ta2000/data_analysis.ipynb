```python
import pandas as pd
import matplotlib.pyplot as plt

# 读取第一个和第十个文件
file_path_1 = 'combined.csv'
file_path_10 = 'dqn_conn0_ep6.csv'

data_1 = pd.read_csv(file_path_1)
data_10 = pd.read_csv(file_path_10)

# 提取第一个文件的数据
steps_1 = data_1['step']
mean_waiting_time_1 = data_1['system_mean_waiting_time']

# 提取第十个文件的数据
steps_10 = data_10['step']
mean_waiting_time_10 = data_10['system_mean_waiting_time']

# 创建第一个文件的折线图
plt.figure(figsize=(10, 6))
plt.plot(steps_1, mean_waiting_time_1, label='mean waiting_time (File 1)', color='blue')
plt.xlabel('Step')
plt.ylabel('System mean waiting_time')
plt.title('System mean waiting_time vs Step (File 1)')
plt.legend()
plt.grid(True)
plt.show()

# 创建第十个文件的折线图
plt.figure(figsize=(10, 6))
plt.plot(steps_10, mean_waiting_time_10, label='mean waiting_time (File 10)', color='green')
plt.xlabel('Step')
plt.ylabel('System mean waiting_time')
plt.title('System mean waiting_time vs Step (File 10)')
plt.legend()
plt.grid(True)
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# 读取一个文件（例如，第一个文件）
file_path = 'dqn_conn0_ep1.csv'
data_sample = pd.read_csv(file_path)

# 选择前100行数据
data_sample_100 = data_sample.head(100)

# 提取step和其他变量
steps_sample = data_sample_100['step']
variables = data_sample_100.drop(columns=['step'])

# 绘制所有变量（除了step）随step变化的折线图
plt.figure(figsize=(12, 8))
for column in variables.columns:
    plt.plot(steps_sample, variables[column], label=column)

plt.xlabel('Step')
plt.ylabel('Values')
plt.title('Variables vs Step (First 100 rows)')
plt.legend(loc='upper left', bbox_to_anchor=(1, 1))
plt.grid(True)
plt.show()
```

```python
import pandas as pd

# Load the uploaded CSV files
file_paths = [
    "dqn_conn0_ep1.csv",
    "dqn_conn0_ep2.csv",
    "dqn_conn0_ep3.csv",
    "dqn_conn0_ep4.csv",
    "dqn_conn0_ep5.csv",
    "dqn_conn0_ep6.csv",
]

# Read the CSV files
dfs = [pd.read_csv(file_path) for file_path in file_paths]

# Add a cumulative step column to each dataframe
for i in range(1, len(dfs)):
    dfs[i]['step'] += dfs[i-1]['step'].max()

# Concatenate all dataframes
combined_df = pd.concat(dfs, ignore_index=True)
# Save the combined dataframe to a new CSV file
combined_df.to_csv("combined.csv", index=False)

```
