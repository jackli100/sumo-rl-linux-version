```python
import pandas as pd

file_paths = 'a2c_conn0_ep2.csv'
steps = pd.read_csv(file_paths)['step']
total_waiting_time = pd.read_csv(file_paths)['system_total_waiting_time']
# 创建折线图
import matplotlib.pyplot as plt
plt.figure
plt.plot(steps, total_waiting_time, label='total_waiting_time')
plt.xlabel('step')
plt.ylabel('total_waiting_time')
plt.title('total_waiting_time')
plt.legend()
plt.show()
```

```python
average_waiting_time = pd.read_csv(file_paths)['system_total_waiting_time'].mean()
print('average_waiting_time:', average_waiting_time)
```
